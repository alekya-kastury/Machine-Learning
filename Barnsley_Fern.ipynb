#Barnsley Fern
Implementation of the Basrnsley Fern using Affine Transformations.

An Affine transformation is a geometric transformation that preserves lines and parallelism but not necessarily angles and distances.

Barnsley Fern uses 4 affine transformations.

$$ùëì(x,y) =
\begin{bmatrix}
a & b \\
c & d \\
\end{bmatrix} \begin{bmatrix}
x \\
y \\
\end{bmatrix} + \begin{bmatrix}
e \\
f \\
\end{bmatrix}
$$

w|	a	|b	|c	|d	|e	|f	|Probability |Portion generated
-|-|-|-|-|-|-|-|-|
f1|	0	    |0	  | 0	   |0.16	 |0	  |0	    |0.01	 |Stem
f2|	0.85	|0.04	| 0.04 | 0.85	 | 0	| 1.60	|0.85	 |Successively smaller leaflets
f3|	0.20	|‚àí0.26|	0.23 |	0.22 | 0	| 1.60	| 0.07	|Largest left-hand leaflet
f4|-0.15	|0.28	| 0.26 | 0.24	 | 0	| 0.44	| 0.07|Largest right-hand leaflet


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Defining the the arrays

```python
A = np.array([
    [0, 0],
    [0, 0.16]
])

B = np.array([
    [ 0.85, 0.04],
    [-0.04, 0.85]
])

C = np.array([
    [0.20, -0.26],
    [0.23,  0.22]
])

D = np.array([
    [-0.15, 0.28],
    [ 0.26, 0.24]
])
```

```python
B0 = np.array([0, 1.6])
C0 = np.array([0, 1])
D0 = np.array([0, 0.44])
```

```python
f1 = lambda x,y: A@np.array([x,y])
f2 = lambda x,y: B@np.array([x,y]) + B0
f3 = lambda x,y: C@np.array([x,y]) + C0
f4 = lambda x,y: D@np.array([x,y]) + D0
```

```python
probabilities=[0.01,0.85,0.07,0.07]
```

```python
import random
X,Y = [], []
x,y = 0, 0
for i in range(50000):
    d = np.random.choice([f1,f2,f3,f4],p=probabilities)
    x, y = d(x,y)
    X.append(x)
    Y.append(y)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(X,Y, s=0.1, color='green')
plt.title('Barnsley Fern')
plt.axis('off')
plt.show()
```
